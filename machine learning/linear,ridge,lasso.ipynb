## Linear Regression, Ridge and Lasso 

```python
#House Pricing dataset 
#imported from inbuilt scikit-learn dataset
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=fetch_california_housing()
```

```python
type(df)
```

```python
df
```

```python
dataset=pd.DataFrame(df.data)
dataset.columns=df.feature_names
dataset.head()
```

```python
dataset['Price']=df.target
```

```python
dataset.head()
```

```python
## Dividing the dataset into independent and dependent features
X=dataset.iloc[:,:-1] ##independent features
y=dataset.iloc[:,-1] ## dependent features
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
## Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
Lin_reg=LinearRegression()
Lin_reg.fit(X_train,y_train)
mse=cross_val_score(Lin_reg,X,y,cv=5, scoring='neg_mean_squared_error')
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
## Linear Regression training model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
Lin_reg=LinearRegression()

mse=cross_val_score(Lin_reg,X_train,y_train,cv=5, scoring='neg_mean_squared_error')
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
## Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
## Ridge Regression training model
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,y_train)
```

```python
## Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X,y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
## Lasso Regression training model
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X_train,y_train)
```

```python
y_pred=lasso_regressor.predict(X_test)
from sklearn.metrics import r2_score

r2_score1=r2_score(y_pred,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
df=load_breast_cancer()
##independent features
X=pd.DataFrame(df['data'],columns=df['feature_names'])
```

```python
X.head()
```

```python
## Dependent feature
y=pd.DataFrame(df['target'],columns=['Target'])
y
```

```python
y['Target'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
params=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model=LogisticRegression(C=100,max_iter=100)
```

```python
model=GridSearchCV(model,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred=model.predict(X_train)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
y_test.shape
```

```python
confusion_matrix(y_pred.shape,y_test.shape)
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_test,y_pred))
```
